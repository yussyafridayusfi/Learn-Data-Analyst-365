```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/bar_chart_data.csv")
df.head()
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df["Brand"], 
        height = df["Cars Listings"],
        color = "rgbwymc" 
      )
plt.xticks(rotation = 45)
plt.show()
```

```python
# resolve issue by Elitsha 365

colors = ["red", "green", "blue", "white", "yellow", "magenta", "cyan"]
plt.figure(figsize=(9, 6))
plt.bar(x=df["Brand"],
        height=df["Cars Listings"],
        color=colors
        )
plt.xticks(rotation=45)
plt.show()
```

```python
colors = ["midnightblue"]
plt.figure(figsize=(9, 6))
plt.bar(x=df["Brand"],
        height=df["Cars Listings"],
        color=colors
        )
plt.xticks(rotation=45)
plt.show()
```

```python
colors = ["midnightblue"]
plt.figure(figsize=(9, 6))
plt.bar(x=df["Brand"],
        height=df["Cars Listings"],
        color=colors
        )
plt.xticks(rotation=45, fontsize = 13)
plt.yticks( fontsize=13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize= 13)

plt.savefig("Used Cars by Brand.png")
plt.show()
```
